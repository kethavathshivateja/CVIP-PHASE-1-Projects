Visualizing Distribution of Key Metrics


```python

# Setting visualisation style
sns.set(style="whitegrid")

# Histograms for numerical columns
df.hist(bins=30, figsize=(12, 10), layout=(5, 2))
plt.tight_layout()
plt.show()

# Box plots for examining potential outliers
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Calculate quartiles
Q1 = df['VALUE'].quantile(0.25)
Q3 = df['VALUE'].quantile(0.75)

# Calculate interquartile range
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['VALUE'] < lower_bound) | (df['VALUE'] > upper_bound)]

# Display summary statistics for 'VALUE'
print("Summary Statistics for 'VALUE' column:")
print(df['VALUE'].describe())

# Display quartile information
print("\nQuartile Information:")
print("First Quartile (Q1):", Q1)
print("Third Quartile (Q3):", Q3)
print("Interquartile Range (IQR):", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Display count and percentage of outliers
outlier_count = len(outliers)
total_count = len(df)
outlier_percentage = (outlier_count / total_count) * 100
print("\nOutlier Information:")
print("Number of Outliers:", outlier_count)
print("Percentage of Outliers:", outlier_percentage)

# Display outliers
print("\nOutliers:")
print(outliers)
```

TOP 10 STORES WITH HIGH SALES ACCORDING TO THE COMPANY

```python
top_stores_by_company = pd.DataFrame(columns=['STORECODE', 'CMP', 'BILL_AMT'])


unique_storecodes = df['STORECODE'].unique()

fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(16, 8))
fig.suptitle('Top 10 Stores with Highest Sales Amounts for Each Company')

# Loop through each unique store code
for i, storecode in enumerate(unique_storecodes[:10]):  # Assuming you want the top 10 stores
    # Filter data for the current store code
    store_data = df[df['STORECODE'] == storecode]

    # Group by company and sum the bill amounts
    top_companies = store_data.groupby('CMP')['BILL_AMT'].sum().reset_index()

    # Sort by bill amounts in descending order and take the top 10
    top_companies = top_companies.sort_values(by='BILL_AMT', ascending=False).head(10)

    top_companies['STORECODE'] = storecode

    # Concatenate the results
    top_stores_by_company = pd.concat([top_stores_by_company, top_companies], ignore_index=True)

    # Plot horizontal bar chart for the current store code
    row, col = i // 5, i % 5
    ax = axs[row, col]
    sns.barplot(x='BILL_AMT', y='CMP', data=top_companies, ax=ax, color='blue')
    ax.set_title(f'Store {storecode}')
    ax.set_xlabel('Bill Amount')
    ax.set_ylabel('Company')

# Adjust layout
plt.tight_layout(rect=[0, 0.10, 1, 0.95])
plt.show()
```

DAILY SALES ANALYSIS FOR EACH STORE

```python
for i, storecode in enumerate(unique_storecodes[:10]): 
    store_data = df[df['STORECODE'] == storecode]
    daily_sales = store_data.groupby("DAY")["BILL_AMT"].sum().reset_index()

    plt.figure(figsize=(20, 5))
    ax = sns.barplot(x=daily_sales['DAY'], y=daily_sales['BILL_AMT'], palette='viridis')
    
    # Add mean line
    mean_sales = daily_sales['BILL_AMT'].mean()
    ax.axhline(mean_sales, color='purple', linestyle='--', label=f'Mean Sales: ${mean_sales:.2f}')
    
    # Add smoothed trend line
    smooth = daily_sales['BILL_AMT'].rolling(window=7).mean()  # 7-day rolling average
    ax.plot(daily_sales['DAY'], smooth, color='orange', linestyle='-', linewidth=2, label='7-Day Rolling Mean')
    
    plt.title("Daily Sales Analysis - Store: " + str(storecode))
    plt.xlabel("Day")
    plt.ylabel("Total Sales")
    plt.grid(axis='y', color='black')
    plt.legend()
    plt.show()
```

TOP 25 CATEGORIES THAT SOLD

```python
# Convert relevant columns to numeric
numeric_columns = ['QTY', 'PRICE', 'VALUE']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Group by 'GRP', sum the numeric columns, and sort by 'VALUE' in descending order
top_items_sales = df.groupby('GRP')[numeric_columns].sum().sort_values("VALUE", ascending=False).head(25)

# Plot the top 25 items by sales value
plt.figure(figsize=(12, 8))
sns.barplot(x='VALUE', y=top_items_sales.index, data=top_items_sales, palette='viridis')
plt.xlabel('Sales Value')
plt.ylabel('Item Group')
plt.title('Top 25 Items by Sales Value')
plt.show()
```

TOP 25 CATEGORIES SOLD IN UNITS

```python
# Convert relevant columns to numeric
numeric_columns = ['QTY', 'PRICE']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Group by 'GRP', sum the numeric columns, and sort by 'QTY' in descending order
top_items = df.groupby('GRP')[numeric_columns].sum().sort_values("QTY", ascending=False).head(25)

# Plot the top 25 items by quantity sold
plt.figure(figsize=(12, 8))
sns.barplot(x='QTY', y=top_items.index, data=top_items, palette='viridis')
plt.title("Number of Units Sold by Top 25 Categories")
plt.xlabel("Number of Units")
plt.ylabel("Categories")
plt.grid(axis='x', color='yellow')

plt.show()
```
